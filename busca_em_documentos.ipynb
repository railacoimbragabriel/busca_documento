```python
df = pd.DataFrame(documents)
df.columns = ["Titulo", "Conteudo"]
df
```

```python
model = "models/embedding-001"
```

```python
def embed_fn (title, text):
  return genai.embed_content(model = model, content = text, title = title, task_type = "RETRIEVAL_DOCUMENT")["embedding"]
```

```python
df["Embeddings"] = df.apply(lambda row: embed_fn(row["Titulo"], row["Conteudo"]), axis = 1)
df
```

```python
# Função para gerar e buscar uma consulta
def gerate_search (query, base, model):

  # Embedding da consulta
  embedding_query = genai.embed_content(model = model, content = query, task_type = "RETRIEVAL_QUERY")

  # Produtos esclares = dot products
  dot_products = np.dot(np.stack(df["Embeddings"]),embedding_query["embedding"])

  # Índice = index
  index =  np.argmax(dot_products) # Buscar o argumento máximo | o que tem mais a ver

  return df.iloc[index]["Conteudo"]
```

```python
query = "Como faço para trocar as marchas no carro?"

trecho = gerate_search (query, df, model)

print(trecho)
```

```python
generation_config = {
    "temperature": 0.5,
    "candidate_count": 1,
    }
```

```python
prompt = f"Reescreva esse texto de uma forma mais descontraída, sem adicionar informações que não façam parte do texto: {trecho}"

model_2 = genai.GenerativeModel(model_name = "gemini-1.0-pro", generation_config = generation_config)
response = model_2.generate_content(prompt)
print(response.text)
```
